## Query OpenAlex for works authored by an institution
This notebook queries the [OpenAlex API](https://docs.openalex.org/api) via its `/works` endpoint for works authored by a institution. It takes an ROR ID as input which is used to filter for works where '`authorships.institutions.ror`' matches the given ROR ID.
The notebook iterates through the given list and displays all DOIs.

```python
# Prerequisites:
import requests         # dependency to make HTTP calls
```

#### Examples of ROR IDs

ID Uni UOS = https://ror.org/04qmmjx98

TIB Hannover = https://ror.org/04aj4c181

```python
ror_id ='https://ror.org/04aj4c181'
```

We use it to query the OpenAlex API for works that specified the ORCID URL within their metadata in the field '`authorships.institutions.ror`'.
 Since the API uses [pagination](https://docs.openalex.org/api/get-lists-of-entities#pagination), we need to loop through all pages to get the complete result set.

```python
# query all works that are connected to orcid
def query_openalex_for_organization2works(ror_id):
    cursor='*'          
    
    while cursor != None:    # Error, Problem letzte Seite
        response = requests.get('https://api.openalex.org/works?filter=authorships.institutions.ror:https://ror.org/04qmmjx98'
                                '&cursor='+cursor
                                )            ####### Suchbegriffe müssen manuell eingegeben werden
        response.raise_for_status()
        result=response.json()
        cursor = result['meta']['next_cursor']
        global total_count
        total_count= result['meta']['count']
        yield result
```

From the resulting list of works we extract and print out title and DOI. 

*Note: works that do not have a DOI assigned, will not be printed. Searchin with a ror id can take a long time*

```python
# from the result pages we get from the OpenAlex API, extract the data about works
def extract_works_from_page(page):
    return [work for work in page.get('results') or []]

# extract DOI from work
def extract_doi(work):
    doi=work.get('ids', {}).get('doi') or ""
    doi_id=doi.replace("https://doi.org/", "") if doi else doi
    title=work.get('display_name', "")
    return doi_id, title

# main Programm
list_of_pages=query_openalex_for_organization2works(ror_id)
for page in list_of_pages or []:
    works=extract_works_from_page(page)
    for work in works or []:
        doi,title=extract_doi(work)
        print(f"{doi},{title}")
```
